```python
import pandas as pd
import io
 
df = pd.read_csv(io.BytesIO(uploaded['churn-bigml-20.csv']))
print(df.shape)
print(df['Churn'])
print(df)
```

**Overall minutes - State wise aggregated- daytime wise, night time wise and international :**

```python
import matplotlib.pyplot as plt

daymins_state_wise = df.groupby('State')['Total day minutes'].sum()
daymins_state_wise_max = df.groupby('State')['Total day minutes'].max()
print("Total day minutes:", daymins_state_wise)
print("Max day minutes:", daymins_state_wise_max)
evemins_state_wise = df.groupby('State')['Total eve minutes'].sum()
evemins_state_wise_max = df.groupby('State')['Total eve minutes'].max()
print("Total evening minutes:",evemins_state_wise)
print("Max eve minutes:", evemins_state_wise_max)
nightmins_state_wise = df.groupby('State')['Total night minutes'].sum()
nightmins_state_wise_max = df.groupby('State')['Total night minutes'].max()
print("Total night minutes:",nightmins_state_wise)
print("Max night minutes:", nightmins_state_wise_max)
intlmins_state_wise = df.groupby('State')['Total intl minutes'].sum()
intlmins_state_wise_max = df.groupby('State')['Total intl minutes'].max()
print("Total intl minutes:",intlmins_state_wise)
print("Max intl minutes:", intlmins_state_wise_max)

Z=pd.concat([daymins_state_wise, evemins_state_wise, nightmins_state_wise, intlmins_state_wise], axis=1)
Z.plot(kind='bar', figsize = (20, 4), ylabel = "Minutes", linewidth = 2)

```
