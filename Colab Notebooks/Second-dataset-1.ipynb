```python
from google.colab import files
 
 
uploaded = files.upload()
```

```python
import pandas as pd
import io
 
df = pd.read_csv(io.BytesIO(uploaded['churn-bigml-20.csv']))
print(df.shape)
print(df['Churn'])
print(df)
```

**Overall churn rate :**

```python
print(df.Churn.sum())
print(df['Churn'].shape[0])
churn_rate = round(df.Churn.sum()/ df['Churn'].shape[0] * 100)
print (churn_rate)
```

**State wise churn rate :**

```python
import matplotlib.pyplot as plt
print(df.groupby('State').Churn.sum())

churn_state_wise = df.groupby('State').Churn.sum()
state_wise_entries = df.State.value_counts()
print(churn_state_wise)
print(state_wise_entries)
print(round((churn_state_wise/state_wise_entries) *100))

plt.figure(figsize=(20,6))
plt.ylabel('Number of people')
churn_state_wise.plot(kind='bar')

plt.figure(figsize=(20,6))
plt.ylabel('Churn Percentage')
plt.xlabel('State')
round((churn_state_wise/state_wise_entries) *100).plot(kind='bar')

plt.show()
```

**Area code wise churn rate:**

```python
print(df.groupby('Area code').Churn.sum())
churn_area_wise = df.groupby('Area code').Churn.sum()
area_code_entries = df.value_counts('Area code')
print(churn_area_wise)
print(area_code_entries)
print(round((churn_area_wise/area_code_entries) *100))
```
