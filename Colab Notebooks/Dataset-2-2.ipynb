```python
import pandas as pd
import io
 
df = pd.read_csv(io.BytesIO(uploaded['January_MyCall_2022.csv']))
print(df.shape)
print(df.describe)
print(df)
```

**BSNL and MTNL not satisfactory calls:**

```python
bsnl_nsatisf=round(df[(df['operator']=='BSNL')&(df['calldrop_category']!='Satisfactory')].shape[0]/df[df['operator']=='BSNL'].shape[0] *100,2)
mtnl_nsatisf=round(df[(df['operator']=='MTNL')&(df['calldrop_category']!='Satisfactory')].shape[0]/df[df['operator']=='MTNL'].shape[0] *100,2)
G2_nsatisf=round(df[(df['operator']=='BSNL') & (df['network_type']=='2G') & (df['calldrop_category']!='Satisfactory')].shape[0]/ df[(df['operator']=='BSNL')&(df['network_type']!='Unknown')].shape[0] *100)
G3_nsatisf=round(df[(df['operator']=='BSNL') & (df['network_type']=='3G') & (df['calldrop_category']!='Satisfactory')].shape[0]/ df[(df['operator']=='BSNL')&(df['network_type']!='Unknown')].shape[0] *100)
G4_nsatisf=round(df[(df['operator']=='BSNL') & (df['network_type']=='4G') & (df['calldrop_category']!='Satisfactory')].shape[0]/ df[(df['operator']=='BSNL')&(df['network_type']!='Unknown')].shape[0] *100)
indoor_nsatisf_bsnl=round(df[(df['operator']=='BSNL')&(df['calldrop_category']!='Satisfactory')&(df['inout_travelling']=='Indoor')].shape[0]/df[df['operator']=='BSNL'].shape[0] *100,2)
overall=df[df['operator']=='BSNL'].value_counts()

print(overall)
print(bsnl_nsatisf,"% are not satisfied with BSNL")
print(mtnl_nsatisf,"% are not satisfied with MTNL")
print("Let's continue the analysis with BSNL")
print(G2_nsatisf,"% are not satisfied with 2G BSNL")
print(G3_nsatisf,"% are not satisfied with 3G BSNL")
print(G4_nsatisf,"% are not satisfied with 4G BSNL")
print(indoor_nsatisf_bsnl, "% are indoors who are not satisfied with BSNL")
```
