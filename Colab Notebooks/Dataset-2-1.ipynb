All data is fictional and obtained from Kaggle.com

```python
from google.colab import files
 
 
uploaded = files.upload()
```

```python
import pandas as pd
import io
 
df = pd.read_csv(io.BytesIO(uploaded['January_MyCall_2022.csv']))
print(df.shape)
print(df)
```

**Operator wise rating average:**

```python
import matplotlib.pyplot as plt
operator_ratings = round(df.groupby('operator')['rating'].mean(),2)
sorted_means = operator_ratings.sort_values(ascending=False)
print("Mean:" , sorted_means)
sorted_means.plot(kind='bar', color='red')
plt.show()

operator_ratings_var = round(df.groupby('operator')['rating'].var(),2)
sorted_means_var = operator_ratings_var.sort_values(ascending=False)
print("Variance:",sorted_means_var)
sorted_means_var.plot(kind='bar', color='green')
plt.show()

operator_ratings_std = round(df.groupby('operator')['rating'].std(),2)
sorted_means_std = operator_ratings_std.sort_values(ascending=False)
print("Standard Deviation:",sorted_means_std)
sorted_means_std.plot(kind='bar', color='blue')
plt.show()

operator_ratings_med = df.groupby('operator')['rating'].median()
sorted_means_med = operator_ratings_med.sort_values(ascending=False)
print("Median:",sorted_means_med)
sorted_means_med.plot(kind='bar', color='violet')
plt.show()
```
